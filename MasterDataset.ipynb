```python
print("Mean:")
print(df.describe().mean())
print("\nMedian:")
print(df.describe().median())
print("\nStandard Deviation:")
print(df.describe().std())
print("\nSkewness:")
print(df.describe().skew())
print("\nKurtosis:")
print(df.describe().kurtosis())
```

```python
print("Volume Statistics:")
print(df['Volume'].describe())
```

```python
plt.hist(df['Volume'], bins=50, edgecolor='black')
plt.title('Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['Close'])
plt.title('Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Close'])
plt.title('Boxplot of Close Price')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(2, 1, 1)
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['Open'], label='Open Price')
plt.title('Close vs Open Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
```

```python
plt.subplot(2, 1, 2)
plt.plot(df['Date'], df['Volume'], label='Volume')
plt.title('Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()

plt.tight_layout()
plt.show()
```
