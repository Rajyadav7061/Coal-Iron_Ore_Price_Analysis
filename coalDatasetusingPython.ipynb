```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/7 to 9/Project/DATASET Iron ore_And Coal Analysis/CoalIndia.csv")
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
print("Mean:")
print(df.describe().mean())
print("\nMedian:")
print(df.describe().median())
print("\nStandard Deviation:")
print(df.describe().std())
print("\nSkewness:")
print(df.describe().skew())
print("\nKurtosis:")
print(df.describe().kurtosis())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'])
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```

```python
print("Volume Statistics:")
print(df['Volume'].describe())
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Volume'], bins=30, edgecolor='black')
plt.title('Distribution of Daily Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Close'])
plt.title('Boxplot of Close Price')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```
